```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('parkinsons.data')
```

```python
df.head()
```

Splitting the data into features and labels ie
features = all columns and not status
target= status

```python
feature_cols=  [val for val in df.columns if val not in ['name','stats']]
features=df.loc[:,feature_cols].values
labels=df.loc[:,'status'].values
```

```python
feature_cols
```

```python
features
```

```python
labels_freq={}
for val in list(set(labels)):
  labels_freq[val]=0
  for label in labels:
      if label==val:
        labels_freq[val]+=1
```

```python
labels_freq
```

```python
scaler=MinMaxScaler((-1,1))
X=scaler.fit_transform(features)
```

```python
y=labels
```

#Splittin the dataset into trainin and testing with 80% training and 20% testing

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
#Intialize the xgbclassifier and train the model
model=XGBClassifier()
model.fit(X_train,y_train,eval_metric="error")
```

```python
y_pred=model.predict(X_test)
```

```python
acc_scr=accuracy_score(y_test,y_pred)*100
```
